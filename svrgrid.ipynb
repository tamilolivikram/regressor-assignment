```python
import pandas as pd
```

```python
import numpy as np
```

```python
dataset=pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset=dataset.astype(int)
dataset
```

```python
dataset.columns
```

```python
ind=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
ind
```

```python
dep=dataset[['charges']]
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ind,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
re=grid.cv_results_
print("The R_score value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
